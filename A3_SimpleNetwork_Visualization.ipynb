### Visualization of the data obtained from simulation

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load data from CSV
path = "HopCountData.csv"
df = pd.read_csv(path)
```

```python
# histogram
plt.figure(figsize=(12, 5))
plt.bar(df['Module'], df['Max'], color='skyblue', edgecolor='black')

# Add labels and title with improved font size and style
plt.xlabel('Module', fontsize=14, fontweight='bold')
plt.ylabel('Max', fontsize=14, fontweight='bold')
plt.title('Maximum Values for Each Module', fontsize=16, fontweight='bold')

# Improve the x-axis ticks for better readability
plt.xticks(rotation=45, ha='right', fontsize=12)

# Add grid for better visual guidance
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()  
plt.show()
```

```python

# Create line chart for Mean
plt.figure(figsize=(12, 5))
plt.plot(df['Module'], df['Mean'], marker='o', linestyle='-', color='coral', markersize=8, linewidth=2)

# Add labels and title with improved font size and style
plt.xlabel('Module', fontsize=14, fontweight='bold')
plt.ylabel('Mean', fontsize=14, fontweight='bold')
plt.title('Mean Hop Count', fontsize=16, fontweight='bold')

# Improve the x-axis ticks for better readability
plt.xticks(rotation=45, ha='right', fontsize=12)

# Add grid for better visual guidance
plt.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()  
plt.show()
```
